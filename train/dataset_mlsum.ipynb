```python
# Korrigiere den Backslash-Fehler und versuche mehrere Varianten
original = "reciTAL\\mlsum"  # aus Zelle 1
candidates = [
    original.replace("\\", "/"),
    original.replace("\\", "/").lower(),
    "mlsum",
]

for name in candidates:
    try:
        ds = datasets.load_dataset(name, "de")
        print("Erfolgreich geladen:", name)
        break
    except Exception as e:
        print(f"Fehler beim Laden '{name}': {e}")
else:
    print("Dataset konnte nicht geladen werden. Prüfe den Dataset-Namen, Internetzugang oder Hugging Face Token.")
```
